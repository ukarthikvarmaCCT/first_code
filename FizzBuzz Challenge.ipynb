# FizzBuzz

#### Challenge 1

Go through all the integers between 1 and 150 inclusive. If the number is divisible by 3, print “Fizz”. If the number is divisible by 5, print “Buzz”. Otherwise, just print the number.
Sample output:

```python
userInput = float(input("Please enter a number between 1 - 150: "))
if userInput % 3 == 0 and userInput % 5 == 0:
    print("FizzBuzz")
elif userInput % 3 == 0:
    print("Fizz")
elif userInput % 5 == 0:
    print("Buzz")
else:
    print(userInput)
```

```python
for num in range(1,151):
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)
```
